```python
dictionary = {}

for idx, frame in reviews.groupby('listing_id'):
    comment = ''
    for j in frame['comments']:
#         print(j)
        comment += str(j)
    dictionary[idx] = comment
```

```python
len(dictionary)
```

```python
reviews = pd.DataFrame({'listing_id': list(dictionary.keys()), 'reviews_clubbed': list(dictionary.values()) })
reviews.head()
```
